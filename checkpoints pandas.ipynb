```python
import pandas as pd
import numpy as np

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthieu', 'Laura', 'Kevin', 'Jonas'],
             'note': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'tentatives': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualifier': ['oui', 'non', 'oui', 'non', 'non', 'oui', 'oui', 'non', 'non', 'oui']}

etiquettes = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(exam_data, index=etiquettes)

print(df)
#question1
print(df.head(3))
#question2
df=df.dropna()
print(df)
#question3
a = ["name","note"]
print(df.loc[:,a])

# Question4 :Création de la nouvelle ligne à ajouter
new = pd.DataFrame({
    'name': 'Suresh',
    'note': 15.5,
    'tentatives': 1,
    'qualifier': 'yes'
}, index=['k'])

# Ajout de la nouvelle ligne au DataFrame
df = pd.concat([df, new])
#Question5
print(df)
df = df.drop('tentatives', axis=1)
print(df)
#question6
df['Succès'] = [1 if note > 10 else 0 for note in df['note']]
print(df)
#question7
df.to_csv('my_data.csv', index=True)
```
